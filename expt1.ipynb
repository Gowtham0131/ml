```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/IRIS[1].csv"
df = pd.read_csv(file_path)
print(df.head())

```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 6))
for i, feature in enumerate(features, 1):
    plt.subplot(1, 4, i)
    sns.scatterplot(
        x=df[feature],
        y=[0]*len(df),  # constant y-axis
        hue=df['species'],
        palette='Set1',
        legend=False
    )
    plt.xlabel(feature)
    plt.yticks([])
    plt.title(f'Scatter plot of {feature} vs constant')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(
    x='sepal_width', y='petal_width',
    hue='species', data=df, palette='Set2'
)
plt.title('sepal_width vs petal_width')

plt.subplot(1,2,2)
sns.scatterplot(
    x='sepal_length', y='petal_length',
    hue='species', data=df, palette='Set2'
)
plt.title('sepal_length vs petal_length')

plt.tight_layout()
plt.show()
```

```python
df_clean = df.drop(columns=['Id'], errors='ignore')

# b & c. Pairplot with KDE on diagonal
sns.pairplot(df_clean, hue='species', diag_kind='kde', palette='Set1')
plt.show()
```
